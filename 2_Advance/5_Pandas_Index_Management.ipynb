```python
# Import Libraries
import os
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Clenup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')
```

## Three Methods to Cover
- df.reset_index()
- df.set_index()
- df.sort_index()

```python
df_no = df[df['job_country'] == 'Norway']
```

```python
df_no.head()
```

```python
df_no.reset_index(inplace=True)
df_no
```

```python
# Setting the index to 'job_index', using this method you loose data.
df_no.set_index('job_index', inplace=True)
```

```python
df_no
```

```python
median_pivot = df_no.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'mean', 'max', 'min'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```
