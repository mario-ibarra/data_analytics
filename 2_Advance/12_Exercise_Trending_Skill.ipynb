```python
# Import Libraries
import os
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Clenup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'], errors='coerce')
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Sort the df by Data Analyst job title and make a copy! Always make a copy when you are going to modify the DataFrame
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA_explode = df_DA.explode('job_skills')

df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no',
                         columns='job_skills',
                         values='job_title_short',
                         aggfunc='size').fillna(0).astype(int)

df_DA_pivot.loc['Total'] = df_DA_pivot.sum(numeric_only=True)
df_DA_pivot= df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot = df_DA_pivot.drop('Total')
df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line', figsize=(12, 6))
plt.title('Top 5 Skills for Data Analyst  per Month')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')

plt.legend(title='Skills', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Reset index to ensure 'job_posted_month_no' is a column
df_DA_pivot = df_DA_pivot.reset_index()

# Convert month number to month name
df_DA_pivot['job_posted_month_no'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(f'2023-{int(x)}-01').strftime('%B'))

# Set month as categorical with correct order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_DA_pivot['job_posted_month_no'] = pd.Categorical(df_DA_pivot['job_posted_month_no'], categories=month_order, ordered=True)
df_DA_pivot = df_DA_pivot.sort_values('job_posted_month_no')

# Set the month name as index for plotting
df_DA_pivot = df_DA_pivot.set_index('job_posted_month_no')
df_DA_pivot.iloc[:, :5].plot(kind='line', figsize=(12, 6))
plt.title('Top 5 Skills for Data Analyst per Month')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.show()
```
